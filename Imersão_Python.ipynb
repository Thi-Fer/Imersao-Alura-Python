**Aula 2 - Importação**

```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/Imersão Python - Tabela de ações.xlsx", sheet_name="Principal")
df_principal.head(10)
```

```python
df_total_acoes = pd.read_excel("/content/Imersão Python - Tabela de ações.xlsx", sheet_name="Total_de_acoes")
df_total_acoes.head(10)
```

```python
df_ticker = pd.read_excel("/content/Imersão Python - Tabela de ações.xlsx", sheet_name="Ticker")
df_ticker.head(10)
```

```python
df_chatgpt = pd.read_excel("/content/Imersão Python - Tabela de ações.xlsx", sheet_name="Chatgpt")
df_chatgpt.head(10)
```

```python
df_principal = df_principal[['Ativo',	'Data',	'Último (R$)',	'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'Valor_Final', 'Var. Dia (%)':'Var_Dia_Pct'}).copy()
df_principal
```

```python
df_principal['Var_Pct'] = df_principal['Var_Dia_Pct']/100
df_principal['Var_Inicial'] = df_principal['Valor_Final']/(df_principal['Var_Pct'] +1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica' : 'Qtde_teorica'}).copy()

df_principal['Variacao_RS'] = (df_principal['Valor_Final'] - df_principal['Var_Inicial']) * df_principal['Qtde_teorica']

df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal
```

```python
df_principal['Qtde_teorica'] = df_principal['Qtde_teorica'].astype(int)
df_principal
```

```python
# Substitua 'df' pelo nome do seu DataFrame e 'O2' pelo nome da sua coluna
df_principal['Resultado'] = pd.Series(['Subiu' if x > 0 else 'Desceu' if x < 0 else 'Estavel' for x in df_principal['Variacao_RS']])
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome da Empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
# Substitua 'df' pelo nome do seu DataFrame e 'S2' pelo nome da sua coluna
df_principal['Cat_idade'] = pd.Series(['Mais de 100' if x > 100 else 'Menos de 50' if x < 50 else 'Entre 50 e 100' for x in df_principal['Idade (anos)']])
df_principal
```

```python
# Calcular os valores desejados
maior = df_principal['Variacao_RS'].max()
menor = df_principal['Variacao_RS'].min()
media = df_principal['Variacao_RS'].mean()
media_subiu = df_principal.loc[df_principal['Resultado'] == 'Subiu', 'Variacao_RS'].mean()
media_desceu = df_principal.loc[df_principal['Resultado'] == 'Desceu', 'Variacao_RS'].mean()

print("Maior:\t\tR$", "{:,.2f}".format(maior))
print("Menor:\t\tR$", "{:,.2f}".format(menor))
print("Média:\t\tR$", "{:,.2f}".format(media))
print("Média Subiu:\tR$", "{:,.2f}".format(media_subiu))
print("Média Desceu:\tR$", "{:,.2f}".format(media_desceu))
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_RS'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_RS'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_RS', text=df_analise_saldo['Variacao_RS'].apply(lambda x: f'R$ {x:,.2f}'), title='Variação reais por Resultado')

fig.show()
```

```python
fig_pie = px.pie(df_analise_segmento, values='Variacao_RS', names='Segmento', title='Variação reais por Segmento')

fig_pie.show()
```

```python
# Supondo que você já tenha o DataFrame df_principal com os dados

# 2. Agrupar os dados por faixa etária (Categ_Idade)
grouped = df_principal.groupby('Cat_idade')

# 3. Calcular a soma da coluna Variacao_RS para cada faixa etária
soma_variacao = grouped['Variacao_RS'].sum()

# 4. Calcular a contagem de empresas para cada faixa etária
quant_empresas = grouped.size()

# 5. Criar um novo DataFrame com os resultados
df_analise_idade = pd.DataFrame({
    'Análise por faixa etária': soma_variacao.index,
    'Variação R$': soma_variacao.values,
    'Quant de Empresas': quant_empresas.values
})

# Exibir os resultados
df_analise_idade
```

```python
fig_bar = px.bar(df_analise_idade, x="Variação R$", y="Quant de Empresas", color="Análise por faixa etária", title="Variação por Faixa Etária")

fig_bar.show()
```
